```python
from bs4 import BeautifulSoup
import requests
import pandas
from pandas import Series,DataFrame
import shutil
```

```python
url='https://www.flipkart.com/mobiles-accessories/pr?sid=tyy&marketplace=FLIPKART'
```

```python
result=requests.get(url)
soup=BeautifulSoup(result.content,'lxml')
```

```python
print(soup)
```

```python
aas=soup.findAll('div',class_='_3BTv9X')
print(aas)
```

```python
image_info = []

for a in aas:
    image_tag = a.findChildren("img")
    image_info.append(('https:'+image_tag[0]["src"], image_tag[0]["alt"]))
```

```python
def download_image(image):
    response=requests.get(image[0],stream=True)
    realname=''.join(e for e in image[1] if e.isalnum())
    
    file=open('C:/Users/Lenovo/Pictures//{}.jpg'.format(realname),'wb')
    
    response.raw.dexode_content=True
    shutil.copyfileobj(response.raw,file)
    del response

'''
def download_image(image):
    response = requests.get(image[0], stream=True)
    realname = ''.join(e for e in image[1] if e.isalnum())
    
    file = open("C://images//bs//{}.jpg".format(realname), 'wb')
    
    response.raw.decode_content = True
    shutil.copyfileobj(response.raw, file)
    del response
'''
```
